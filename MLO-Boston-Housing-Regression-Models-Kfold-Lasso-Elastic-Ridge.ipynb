```python
data = pd.read_csv("C:\\Users\\Aravinthan\\Desktop\\MLO\\data\\Boston.CSV")
```

```python
data.head()
```

```python
x = data.iloc[:,:12].values
y = data.iloc[:,12].values
```

```python
kf = KFold(n_splits=5)
```

```python
model = LinearRegression()
```

```python
list(kf.split(data))
```

```python
x = data.iloc[:,:12].values
y = data.iloc[:,12].values
```

```python
RMSE = []

for fold in list(kf.split(data)):
    model.fit(x[fold[0],:], y[fold[0]])
    y_pred = model.predict(x[fold[1],:])
    RMSE.append(np.sqrt(mean_squared_error(y_pred, y[fold[1]])))

print(RMSE)
```

```python
cvals = -cross_val_score(model,x,y,cv=10,scoring="neg_mean_squared_error")
```

```python
cvals = np.sqrt(cvals)
cvals
```
